```python
Hesap_Makinası = """

*********************

İşlem listesi

1.Toplama işlemi
2.Çıkarma işlemi
3.Çarpma işlemi
4.Bölme işlemi
5.Kare bulma işlemi
6.Karekök bulma işlemi

*********************

"""

print (Hesap_Makinası)

while True:
    

    Değer = input("Bir İşlem Numarası girin, Çıkmak için q ya basın:")
    
    if Değer =="q":
        print("Programdan çıkılıyor")
        break


    elif Değer == "1":
        sayı1 = int(input("Toplama işlemi için 1. sayıyı girin:"))
        sayı2 = int(input("Toplama işlemi için 2. sayıyı girin:"))
        print("{} sayısı ile {} sayısının toplamı {} dır.".format(sayı1,sayı2,sayı1+sayı2))
    elif Değer == "2":
        sayı1 = int(input("Çıkarma işlemi için 1. sayıyı girin:"))
        sayı2 = int(input("Çıkarma işlemi için 2. sayıyı girin:"))
        print("{} sayısı ile {} sayısının farkı {} dır.".format(sayı1,sayı2,sayı1-sayı2))
    elif Değer == "3":
        sayı1 = int(input("Çarpma işlemi için 1. sayıyı girin:"))
        sayı2 = int(input("Çarpma işlemi için 2. sayıyı girin:"))
        print("{} sayısı ile {} sayısının Çarpımı {} dır.".format(sayı1,sayı2,sayı1*sayı2))
    elif Değer == "4":
        sayı1 = int(input("Bölme işlemi için 1. sayıyı girin:"))
        sayı2 = int(input("Bölme işlemi için 2. sayıyı girin:"))
        print("{} sayısı ile {} sayısının bölümü {} dır.".format(sayı1,sayı2,sayı1/sayı2))
    elif Değer == "5":
        sayı1 = int(input("Kare bulma işlemi için 1. sayıyı girin:"))
        print("{} sayısının karesi {} dır.".format(sayı1,sayı1**2))
    elif Değer == "6":
        sayı1 = int(input("Karekök bulma işlemi için 1. sayıyı girin:"))
        print("{} sayısının karekökü {} dır.".format(sayı1,sayı1**0.5))
    else:
        print("Geçersiz işlem")
        print("Aşağıdaki işlemlerden birini giriniz", Hesap_Makinası)

```
